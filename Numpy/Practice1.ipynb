```python
# step must be a number, not a tuple
dim_mat = np.arange(48).reshape(4,3,4)
dim_mat2 = np.arange(48,96).reshape(4,3,4)
print(dim_mat)


# MSE

error = 1/48 * np.sum(np.square(dim_mat-dim_mat2))


#     or pro method 

error2 = np.mean(np.square(dim_mat-dim_mat2))
print(error)
print(error2)
```
