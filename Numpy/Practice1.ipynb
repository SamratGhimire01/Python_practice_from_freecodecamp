```python
mat=np.random.randint(10,50,(4,5))
data1=np.random.randint(5,10,(3,4))@mat
print(data1)
print(data1[1:2,2:3])
print(data1[1:,3:])
print(data1[2,4])
```

```python
print(data1.max(axis=0).reshape(5,1))
print(data1.max(axis=1))
print(data1.T)
```

```python
# step must be a number, not a tuple
dim_mat = np.arange(48).reshape(4,3,4)
dim_mat2 = np.arange(48,96).reshape(4,3,4)
print(dim_mat)


# MSE

error = 1/48 * np.sum(np.square(dim_mat-dim_mat2))


#     or pro method 

error2 = np.mean(np.square(dim_mat-dim_mat2))
print(error)
print(error2)
```
