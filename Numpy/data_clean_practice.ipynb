```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dataset/messy_ecommerce_sales_data.csv",delimiter=",")
df.describe()
df.info()
df.isna().sum()
```

```python

df['Price'] = pd.to_numeric(df['Price'].astype(str),errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'],errors='coerce')
df['Order_Date'] = pd.to_datetime(df['Order_Date'],errors='coerce')
df['Order_Date'] = df['Order_Date'].fillna(df['Order_Date'].mode()[0])
df['Price']=(df['Price'].fillna(df['Price'].mean())).round(2)
df['Quantity']=(df['Quantity'].fillna(1)).abs()
df['Total'] =( df['Price'] * df['Quantity']).round(2)
df.to_csv('updated_file.csv',sep=",")
print(df['Order_Date'].to_string())
```
